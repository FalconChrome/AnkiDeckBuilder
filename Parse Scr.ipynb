```python
def get_df(table):
    ''' Make DataFrame of the list copied from dictionary in pdf.
    
    Potentially with cli prompting in non-straight cases
    prompt = input("""==> Copypaste from table of the format:
    - [B]: [...1 åŠ  jiÄ ï¼ˆåŠ¨è¯ï¼‰ Ð´Ð¾Ð±Ð°Ð²Ð¸Ñ‚ÑŒ...]
    - [C]: [1 2 3 4...27 28 åŠ  æ²¹...]
    ==> Boxing method? [B/c]: """)[0]'''
    
    import numpy as np
    
    for i in range(1, len(table)):
        if table[i].isdecimal():
            L = i
            break
    
    prompt = 'c' if i == 1 else 'b'
    
    match prompt:
        case 'b':
            N = len(table)//L
            rows = np.array(table).reshape(N, L).tolist()
        case 'c':
            N = sum(1 for _ in (filter(str.isdecimal, table)))
            L = len(table)//N
            rows = np.array(table).reshape(L,N).transpose().tolist()
        case _:
            raise KeyError(f"Unreckognized option {prompt}")

    
    import pandas as pd
    
    columns = ["id", "hanzi", "pinyin", "pos", "trans"]
    
    # df['id'] = df.index
    if (len(columns) - L):
        if prompt:=input('''==> Column names to retain? 
      ==> ([I]d [H]anzi [P]intyin [Po]S [T]ranslation) or (1 2 3) 
      ==> '''):
            column_prompt = [c[:2].lower() for c in prompt.split()]
            
            
            if all(str.isdecimal, column_prompt):
                columns = [columns[int(i)] for i in column_prompt]
        
            elif all(columns.__contains__, column_prompt):
                columns = [(next(filter(lambda c: c.startswith(p), columns), None)) 
                               for p in column_prompt]
                columns = list(filter(bool, columns))
            else: 
                columns = column_prompt

    df = pd.DataFrame(rows, columns=columns)

    if "id" in df.columns and df["id"].is_unique:
        df.set_index('id', inplace=True)

    return df
# rows
```

```python
import random
def genid():
    return random.randrange(1 << 30, 1 << 31)
```

```python
# %pip install genanki
import genanki
doll = genanki.Model(
  2055766351,
  'Simple with PoS',
  fields=[
    {'name': 'Id'},
    {'name': 'Hanzi'},
    {'name': 'Pinyin'},
    {'name': 'Part of Speech'},
    {'name': 'Translation'}
  ],
  templates=[
    {
      'name': 'Card {{Id}}',
      'qfmt': '''
<div style='font-family: "Arial"; font-size: 80px;'>{{Hanzi}}</div>''',
      'afmt': '''{{FrontSide}}

<hr id=answer>

<div style='font-family: "Arial"; '>{{Pinyin}}</div>
<div style='font-family: "Arial"; '>{{Part of Speech}}</div>
<div style='font-family: "Arial"; '>{{Translation}}</div>
''',
    },
  ],
  css='''.card {
    font-family: arial;
    font-size: 60px;
    text-align: center;
    color: black;
    background-color: white;
}''')
```

```python
def save_deck(df, file, name, deckid=None):
    if deckid is None:
        deckid = genid()
    
    ch_deck = genanki.Deck(
      deckid,
      name)
    
    # for id in df1.index:
    #     rec = tuple(df1.loc[id])
    #     ch_deck.add_note(genanki.Note(
    #   model=chineese1,
    #   fields=[str(id), *rec]))
    
    for id in df.index:
        rec = tuple(df.loc[id])
        ch_deck.add_note(genanki.Note(
      model=doll,
      fields=[str(id), *rec]))
    
    genanki.Package(ch_deck).write_to_file(file)
```

```python
import argparse

if __name__ == "__main__":
    parser = argparse.ArgumentParser(prog='Anki Deck Builder', 
                                     description="Parsing dictionary tables to anki entries",
                                     epilog="Made for Chinese (MIPT LPR 02/25/25ðŸ„¯)")
    parser.add_argument('-d', '--deck', help="Title of the created deck", nargs='?')
    parser.add_argument('text', help="Input text copied from dictionary table")
    parser.add_argument('-f', '--file', help="Filename to save the deck", nargs='?', default='output.apkg')
    
    args = parser.parse_args()
    
    text = args.text
    table = text.splitlines()
    df = get_df(table)
    save_deck(df, args.file, args.name, args.deckid)
    
```
